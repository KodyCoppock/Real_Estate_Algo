```python
def sma(countryNominaldf, period):
    j = 0
    summ = 0
    while j < period * 4:
        summ = summ + countryNominaldf.at[j, "Value"]
        j = j + 1
    avg = summ / (period * 4)
    return avg
```

```python
def ema(countrydf, period):
    j = (period * 4)
    count = 1
    smaVar = sma(countrydf, period)
    smooth = 2 / ((period * 4) + 1)
    ema = ([[""] * 3] * (len(countrydf.index) - (period * 4)))
    emadf = pd.DataFrame(ema, columns = ['Country', "Time", 'Value'])
    emadf.at[0, "Value"] = smaVar
    while j < len(countrydf.index) - 1:
            emadf.at[count, "Value"] = ((countrydf.at[j, "Value"] - emadf.at[count - 1, "Value"]) * smooth) + emadf.at[count - 1, "Value"]
            emadf.at[count, "Country"] = countrydf.at[j, "Country"]
            j = j + 1
            count = count + 1
    return emadf
```

```python
def macd(countrydf, shortPeriod, longPeriod, signalPeriod):
    short = ema(countrydf, shortPeriod)
    long = ema(countrydf, longPeriod)
    mac = ([[""] * 3] * len(long.index))
    macdLine = pd.DataFrame(mac, columns = ['Country', 'Value', 'Signal'])
    j = 0
    k = signalPeriod * 4
    macdLine.at[0, "Country"] = long.at[0, "Country"]
    diff = (longPeriod - shortPeriod) * 4
    while j < len(long.index):
        macdLine.at[j, "Value"] = (short.at[j + diff, "Value"] - long.at[j, "Value"])
        macdLine.at[j, "Country"] = long.at[j, "Country"]
        j = j + 1
    lineSignal = ema(macdLine, 2)
    temp = 0
    while k < len(long.index):
        macdLine.at[k, "Signal"] = lineSignal.at[temp, "Value"]
        k = k + 1
        temp = temp + 1
    return macdLine
```

```python
def macdDiff(df):
    j = 0
    macdDiff = []
    while j < len(df.index):
        if type(df.at[j, "Signal"]) == str:
            macdDiff.append(None)
            j = j + 1
        else:
            macdDiff.append(df.at[j, "Value"] - df.at[j, "Signal"])
            j = j + 1
    return macdDiff
```

```python
def plotMacdDiff(df):
    j = 0
    k = 0
    l = 0
    m = 0
    zero = []
    date = []
    while j < len(df):
        zero.append(0)
        j = j + 1
    # Data
    while k < len(df):
        if df[k] == None:
            df[k] = 0
        k = k + 1
    while l < len(df):
        date.append(m)
        m = m + 1
        l = l + 1
    x = np.asarray(date)
    y = np.asarray(df)
 
    # Create a color if the group is "B"
    my_color=np.where(y>=0, 'green', 'red')
 
    # The vertival plot is made using the hline function
    # I load the seaborn library only to benefit the nice looking feature
    plt.vlines(x=x, ymin=0, ymax=y, color=my_color, alpha=0.4)
    plt.scatter(x, y, color=my_color, s=1, alpha=1)
 
    # Add title and axis names
    plt.title("Differece between MACD and Signal Line", loc='left')
    plt.xlabel('Quarter')
    plt.ylabel('Indicator (MACD - Signal Line)')
    return
```

```python
def taxAdjustedPR(prdf, effTaxRate):
    adjustedPR = prdf.copy()
    j = 0
    while j < len(adjustedPR.index):
        adjustedPR.at[j, "Value"] = adjustedPR.at[j, "Value"] - (adjustedPR.at[j, "Value"] * (float(effTaxRate) / 100))
        j = j + 1
    return adjustedPR
```

```python
def bulls(country1, *args):
    sor = ([[""]] * 1000)
    sort = pd.DataFrame(sor, columns = ['Country'])
    temp= 0
    if len(country1.index) > 40:
        country1MACD = macd(country1, 5, 7, 3)
        country1MACD = macdDiff(country1MACD)
        if country1MACD[len(country1MACD)-1] > 0:
            sort.at[temp, "Country"] = country1.at[0, 'Country']
            temp = temp + 1
    for arg in args:
        if len(arg.index) > 40: 
            argMACD = macd(arg, 5, 7, 3)
            argMACD = macdDiff(argMACD)
            if argMACD[len(argMACD)-1] > 0:
                sort.at[temp, "Country"] = arg.at[0, 'Country']
                temp = temp + 1
    sort = dropEmpty3(sort)
    return sort
```
